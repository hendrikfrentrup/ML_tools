```python
import numpy as np
from pandas import read_csv

data_file='../data/ip_vuln_id.csv'

ip_vuln_data = read_csv(data_file, 
                        delimiter=',', 
                        header=None, 
                        names=['ip','vuln_id'], 
                        dtype={'ip':'string', 'vuln_id':np.int64}
                       )
```
